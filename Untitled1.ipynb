```python
dum=pd.get_dummies(df['MaritalStatus'],drop_first=True)
dum1=pd.get_dummies(df['Gender'],drop_first=True)
dum.head()
```

```python
dum1.head()
```

```python
df=pd.concat([df,dum,dum1],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(df,train_size=0.7,random_state=100)
```

```python
pro = {'TM195': 1, "TM498": 2,"TM798":3} 
df_train.Product=[pro[item] for item in df_train.Product]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
var=['Income','Miles']
df_train[var]=scaler.fit_transform(df_train[var])
```

```python
df_train.tail(10)
```

```python
y_train=df_train['Fitness']
x_train=df_train[['Product','Education','Age','Usage','Income','Miles','Single','Male']]
x_train.head()
```

```python
import statsmodels.api as sm
x_train_=sm.add_constant(x_train)
lr=sm.OLS(y_train,x_train_).fit()
lr.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x=x_train.drop('Education',axis=1)
x_train_1=sm.add_constant(x)
x_train_1.head()
```

```python
lr=sm.OLS(y_train,x_train_1).fit()
lr.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x1=x_train.drop(['Education','Usage','Age'],axis=1)
x_train_2=sm.add_constant(x1)
x_train_2.head()
```

```python
lr=sm.OLS(y_train,x_train_2).fit()
lr.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = x1.columns
vif['VIF'] = [variance_inflation_factor(x1.values, i) for i in range(x1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
